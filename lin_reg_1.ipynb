```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.optimize import curve_fit

y = np.array([12.19, 8.41, 14.68, 8.64, 32.94, 22.61, 45.92, 23.63, 18.59, 36.22, 50.10, 46.22, 23.63, 47.30, 40.03, 56.53, 38.41, 51.47, 6.29, 35.41, 67.79, 74.21, 79.12, 45.10])
x = np.linspace(0, 1, num = 24)
plt.scatter(x, y);
```

```python
X = list(zip(np.ones(24), x, x*x))
W = np.linalg.lstsq(X, y, rcond=None)[0]
```

```python
W
```

```python
def f(x, w): 
    return w[0] + w[1]*x + w[2]*x*x
```

```python
f(x, W)
```

```python
plt.scatter(x, y, label='train set')
plt.plot(x, f(x, W), label='quad lstsq')
plt.legend();
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y, f(x, W))
```

```python
X = list(zip(np.ones(24), x, x*x,x*x*x))
W = np.linalg.lstsq(X, y, rcond=None)[0]
```

```python
W
```

```python
def f(x, w): 
    return w[0] + w[1]*x + w[2]*x*x + w[3]*x*x*x
```

```python
f(x, W)
```

```python
plt.scatter(x, y, label='train set')
plt.plot(x, f(x, W), label='quad lstsq')
plt.legend();
```

```python
mean_squared_error(y, f(x, W))
```

```python
X = list(zip(np.ones(24), np.exp(x)))
W = np.linalg.lstsq(X, y, rcond=None)[0]
```

```python
W
```

```python
def f(x, w): 
    return w[0] + w[1]*np.exp(x)
```

```python
f(x, W)
```

```python
plt.scatter(x, y, label='train set')
plt.plot(x, f(x, W), label='quad lstsq')
plt.legend();
```

```python
mean_squared_error(y, f(x, W))
```

1. MSE для полинома второй степени: 212,5
2. MSE для полинома третьей степени: 202,5
3. MSE для экспоненты: 218,2

MSE - это среднеквадратичная ошибка или среднеквадратичное отклонение, то наилучшим вариантом является полином третьей степени.
